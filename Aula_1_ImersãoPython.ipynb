```python
import pandas as pd
```

```python
# df = data frame
# pd.read irá ler o arquivo que estarão os dados que serão analisados
df = pd.read_csv('https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv')
```

```python
# armazenar os dados em duas variáveis distintas
linhas, colunas = df.shape[0], df.shape[1]

# print dos números de linhas e colunas
print('linhas:', linhas)
print('colunas:', colunas)
```

```python
# colunas do data frame
df.columns
```

```python
# tradução das colunas para o português
colunas_traduzidas = {
    'work_year': 'ano',
    'experience_level': 'nivel_experiencia',
    'employment_type': 'tipo_emprego',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda_salario',
    'salary_in_usd': 'salario_em_usd',
    'employee_residence': 'residencia_funcionario',
    'remote_ratio': 'remoto',
    'company_location': 'localizacao_empresa',
    'company_size': 'tamanho_empresa'
}

df = df.rename(columns=colunas_traduzidas)
print(df.columns)
```

```python
df.head()
```

```python
# mostra os valores totais que as categorias da coluna 'nivel_experiencia' aparecem
df['nivel_experiencia'].value_counts()
```

```python
df['cargo'].value_counts()
```

```python
df['tipo_emprego'].value_counts()
```

```python
# combinação de tipo_emprego com cargo
df[['tipo_emprego', 'cargo']].value_counts()
```

```python
df['remoto'].value_counts()
```

```python
df['tamanho_empresa'].value_counts()
```

```python
# renomear valores dentro da coluna 'nivel_experiencia'.
df['nivel_experiencia'] = df['nivel_experiencia'].replace({
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
})
df['nivel_experiencia'].value_counts()
```

```python
df['tipo_emprego'] = df['tipo_emprego'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
})
df['tipo_emprego'].value_counts()
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
})
df['remoto'].value_counts()
```

```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'M': 'Médio',
    'L': 'Grande',
    'S': 'Pequeno'
})
df['tamanho_empresa'].value_counts()
```
