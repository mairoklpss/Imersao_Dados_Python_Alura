```python
# importação de biblioteca para valores nulos
import numpy as np
```

```python
# ----CRIAÇÃO DE UM NOVO DATAFRAME DO 0----
df_salarios = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Dani', 'Maira'],
    # np.nan = valor nulo
    'salario': [4000, np.nan, 7500, np.nan, 100000]
})

'''
preenchimento das celulas nulas com a MÉDIA (mean) dos valores do campo 'salario':
cria-se um novo campo (salario_media) que irá receber os campos nulos preenchidos
.fillna(df_salarios('salario'.mean().round2)) método que irá preencher (fill) as celulas com a MÉDIA
dos salários com até duas casas decimais (round(2)).
'''

df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

'''
mesma lógica, mas preenchendo com a MEDIADA (median)
mediana se torna melhor, pois pode ser mais realista
se caso houver um salario muito alto, a média pode retornar valores equivocados
'''

df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
# ----CRIAÇÃO DO DATAFRAME TESTE----
df_temperaturas = pd.DataFrame({
    'dia': ['segunda',  'terça', 'quarta', 'quinta', 'sexta'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

# ffill (forwardfill) é um método do pandas que preenche as células vazias com o valor anterior.
df_temperaturas['preenchido_ffill'] = df_temperaturas['temperatura'].ffill()
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    'dia': ['segunda',  'terça', 'quarta', 'quinta', 'sexta'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

# bfill (backfill) é um método do pandas que preenche as células vazias com o valor posterior.
df_temperaturas['preenchido_bfill'] = df_temperaturas['temperatura'].bfill()
df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Dani', 'Maira'],
    'cidade': ['São Paulo', np.nan, 'Tianguá', np.nan, 'Arapá']
})

# fillna preenche as células vazias com o que for escrito dentro dos parênteses.
# no caso, está preenchido com "Não Informado".
df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('Não Informado')
df_cidades
```

```python
# importação do matplotlib, em específico o método pyplot
import matplotlib.pyplot as plt
```

```python
# usando matplotlib para definir o tamanho do gráfico em barras
plt.figure(figsize=(10, 8))
# código em seaborn
sns.barplot(data=df_limpo, x='nivel_experiencia', y='salario_em_usd')
# título gráfico
plt.title('Sálario Médio por Nível de Experiência')
# legenda do eixo x
plt.xlabel('Nível de Experiẽncia')
# legenda dp eixo y
plt.ylabel('Salário Médio Anual (em dólar)')
# esse comando mostra o gráfico bonitinho.
plt.show()
```

```python
# ordenar o gráfico do maior valor para o menor valor.
# ascending=False: maior - menor
# ascending=True: menor - maior
df_limpo.groupby('nivel_experiencia')['salario_em_usd'].mean().sort_values(ascending=False).round(2)
```

```python
# colocando o código anterior em uma variável para usá-la no matplotlib.
# COLOCAR .index NO FINAL.
dados_maior_menor = df_limpo.groupby('nivel_experiencia')['salario_em_usd'].mean().sort_values(ascending=False).round(2).index
dados_maior_menor
```

```python
# usando matplotlib para definir o tamanho do gráfico em barras
plt.figure(figsize=(10, 8))
# código em seaborn
sns.barplot(data=df_limpo, x='nivel_experiencia', y='salario_em_usd', order=dados_maior_menor)
# título gráfico
plt.title('Sálario Médio por Nível de Experiência')
# legenda do eixo x
plt.xlabel('Nível de Experiẽncia')
# legenda dp eixo y
plt.ylabel('Salário Médio Anual (em dólar)')
# esse comando mostra o gráfico bonitinho.
plt.show()
```

```python
# crianção de variável para usá-la no matplotlib.
# menor para o maior.
dados_menor_maior = df_limpo.groupby('nivel_experiencia')['salario_em_usd'].mean().sort_values(ascending=True).round(2).index
dados_menor_maior
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(data=df_limpo, x='nivel_experiencia', y='salario_em_usd', order=dados_menor_maior, palette= 'Set2', hue= 'nivel_experiencia')

plt.title('Salário Médio por Níveis de Experiência')
plt.xlabel('Nível de Experiẽncia')
plt.ylabel('Salário Médio Anual (em dólar)')

plt.show()
```

```python
plt.figure(figsize=(10,5))
# CRIAÇÃO DE UM HISTOGRAMA
# bins: quantidade de barras
# kde: linha que contorna as barras
sns.histplot(df_limpo['salario_em_usd'], bins = 50, kde = True)
plt.title('Distribuição de Salários em USD')
plt.xlabel('Salário em USD')
plt.ylabel('Frequência')
plt.show()
```

```python
plt.figure(figsize=(8,4))
# CRIAÇÃO DE UM DIAGRAMA DE CAIXA
# usando dados dos salários em USD.
sns.boxplot(x=df_limpo['salario_em_usd'])
plt.title('Boxplot de Salários em USD')
plt.xlabel('Salário em USD')
plt.show()
```

```python
ordem_senioridade = ['Junior', 'Pleno', ' Senior', 'Executivo']
plt.figure(figsize=(8,5))
# CRIAÇÃO DE BOXPLOT COM DOIS EIXOS
# analisando os salários em USD de cada nível de experiência.
sns.boxplot(x='nivel_experiencia', y='salario_em_usd', data=df_limpo, order=dados_menor_maior)
plt.title('Distribuição Salarial por Senioridade')
plt.xlabel('Nível de Senioridade')
plt.ylabel('Salário em USD')
plt.show()
```

```python
# colocando cores por meio do parametro 'palette=' e 'hue='.
sns.boxplot(data= df_limpo, x='nivel_experiencia', y='salario_em_usd', order=dados_menor_maior, palette= 'husl', hue= 'nivel_experiencia')
plt.title('Distribuição Salarial por Senioridade')
plt.xlabel('Nível de Senioridade')
plt.ylabel('Salário em USD')
plt.show()
```

```python
# importação da biblioteca plotly.express para gráficos interativos
import plotly.express as px
```

```python
# ---- CRIAÇÃO DE GRÁFICO EM BARRAS INTERATIVO ----
df_grouped_salary = df_limpo.groupby('nivel_experiencia')['salario_em_usd'].mean().sort_values(ascending=False).round(2).reset_index()

fig = px.bar(
    df_grouped_salary,
    x='nivel_experiencia',
    y='salario_em_usd',
    title='Média Salarial Anual em USD por Nível de Senioridade',
    labels={'nivel_experiencia': 'Nível de Senioridade', 'salario_em_usd': 'Média Salarial Anual (USD)'},
    color='nivel_experiencia'
)

fig.update_layout(xaxis_title='Nível de Senioridade', yaxis_title='Média Salarial Anual (USD)')
fig.show()
```

```python
# ---- CRIAÇÃO DE GRÁFICO DE PIZZA INTERATIVO ----
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()

remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values= 'quantidade',
             title= 'Proporção dos tipos de trabalho'
             )

fig.show()
```

```python
# ---- CRIAÇÃO DE GRÁFICO DE ROSCA INTERATIVO ----
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()

remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values= 'quantidade',
             title= 'Proporção dos tipos de trabalho',
             hole= 0.5
             )

fig.update_traces(textinfo= 'percent+label')
fig.show()
```

```python
# filtrei as linhas onde o cargo é apenas 'Data Scientist' e coloquei em uma variavel.
df_data_scientist = df_limpo[df_limpo['cargo'] == 'Data Scientist']
# coloquei em ordem descrescente a media dos salarios de acordo com cada país da variavel que criei: 'df_data_scientist'
# e armazenei em outra variável.
df_salario_scientist = df_data_scientist.groupby('residencia_funcionario')['salario_em_usd'].mean().sort_values(ascending = False).round(0).reset_index()
df_salario_scientist
```

```python
fig = px.bar(
    df_salario_scientist,
    x = 'residencia_funcionario',
    y = 'salario_em_usd',
    title = 'Média Salarial Anual em USD de Data Scientists por País',
    labels={'residencia_funcionario': 'País', 'salario_em_usd': 'Salário Anual (em USD)'},
    color='residencia_funcionario'
)

fig.show()
```
